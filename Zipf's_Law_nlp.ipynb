# Zipf's Law

##### States that given some corpus Of natural language utterances The frequency of any word Is inversely proportional  To its rank In the Frequency table.

```python
from nltk.corpus import brown
```

```python
Words = brown.words()
print(Words)
```

```python
len(Words)
```

# Over 1 Millon Tokens

```python
from collections import Counter
pun = set((',','.','--','-','!','?',':',';','``',"''",'(',')','[',']'))
```

```python
word_list = (x.lower() for x in Words if x not in pun)
```

```python
token_counts = Counter(word_list)
```

```python
token_counts.most_common(20)
```

```python
from operator import itemgetter
import pandas as pd
frequency = token_counts
rank = 1
column_header = ['Rank', 'Frequency']
df = pd.DataFrame( columns = column_header )
collection = sorted(frequency.items(), key=itemgetter(1), reverse = True)
```

# itemgetter is a pointer

```python
for word , freq in token_counts.most_common(20):
    df.loc[word] = [rank, freq]
    rank = rank + 1
print (df)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(20,20))  #to increase the plot resolution
plt.ylabel("Frequency")
plt.xlabel("Words")
plt.xticks(rotation=90)    #to rotate x-axis values

for word , freq in collection[:10]:
    plt.bar(word, freq)    
plt.show()
```
